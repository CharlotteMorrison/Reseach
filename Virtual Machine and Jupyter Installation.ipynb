## Install a virtual machine

### Download files
1.  Download VirtualBox https://www.virtualbox.org
2.  Download Ubuntu Linux ISO distribution https://ubuntu.com/desktop

### Install VirtualBox
Double click the virtualbox.exe file to run the installation. 
The virtualbox.exe file name will look similar to this one: _VirtualBox-6.0.8-130520-Win.exe_

### Set up virtual environment
1. Run VirtualBox.
2. When it is open, click the New symbol and fill out the information:
>Name:  		 _whatever you choose_<br>
>Machine Folder: _location of your installation_<br>
>Type:  		 _Linux_<br>
>Version: 		 _Ubuntu (64-bit) Linux_<br>
><br>
>(click next)<br>
3. Allocate from 2-4 GB of ram (1024-4076 MB), should be less than half of your system's resources. (click next)
4. Choose to create a virtual hard disk. (click next)
5. Choose VDI hard disk file type. (click next)
6. Choose dynamically allocated hard disk. (click next)
7. For the file size and location allocate at least 10GB, but 15-20 is better if you have the space.
8. Click Create.

### Install Ubuntu Linux 
1. Select your virtual machine and click the green start arrow.
2. Select the start up disk by finding your ISO (navigate to downloads unless you saved it elsewhere).  The file name will look like _ubuntu-18.04.2-desktop-amd64.iso_ 
3. Click Start.
4. Choose install Ubuntu on the welcome screen.
5. Select download updates while installing Ubuntu and then continue.
6.  The remainder of the steps are straightforward except installation type:
Erase disk and install Ubuntu __will not__ impact your computer's primary operating system.  It is referening to the virtual disk you set up earlier.

You will need to restart your computer once the installation is complete.

### Extra Stuff
To make the screen full size within the window, open the devices tab and choose "Insert guest additions CD image" this will run some commands, press enter to finish. The screen will automatically resize.

To enable copy/paste and drag and drop between the host and the guest:
1. In VirtualBox manager select settings --> general --> advanced (tab) 
2. Choose Bidirectional from the dropdown menus for Shared Clipboard and Drag'n'Drop.  
3. Restart virtual machine.

## Install Python and Jupyter Notebook

### Set up Python
Update package index<br>
`sudo apt update`

Install pip andy python header files<br>
`sudo apt install python3-pip python3-dev`

### Create a Virtual Environment
Jupyter will be installed to this environment.

Update pip and install virtualenv.<br>
`sudo -H pip3 install --upgrade pip`<br>
`sudo -H pip3 install virtualenv`

Create a directory for the project and cd into it.<br>
`mkdir ~/deep_rl`<br>
`cd ~/deep_rl`

Create a virtual environment, then activate it.<br>
`virtualenv deep_rl_env`<br>
`source deep_rl_env/bin/activate`<br>
If you did this correctly the virtualenv name will be prepended on the prompt in parenthesis.

### Install and run Jupyter 
(do not use pip3 inside of the virtual environment, use just pip)<br>
`pip install jupyter`<br>
`jupyter notebook`

It will open in your default browser.
