### Observations fom environment

Each step returns four values:<br>

- observation (object): this is specific to the environment and is the observation of the current state.
- reward (float): the reward from the previous action. 
- done (boolean): if the environment needs to be reset.  This would indicat that the episode has termininated (lost the game, failed the task, hit an end condition).
- info (dict): for debugging- not for evaluations.

The process starts with reset() returning the initial observation, at each step, an action is chosen and the environment returns an observation and reward.

```python
import gym
env = gym.make('CartPole-v0')
for i in range(20):
    observation = env.reset()
    for t in range (100): 
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print('Episode Finishd afer {} timesteps.'.format(t+1))
            break
env.close()
```

### Get the actions for the specific environment
___
Each environment has an action_space and an observation_space.<br>
The action space is discrete (0,1) <br>
The observation space is a n-dimensional box, valid observations will be an array of 4 numbers.<br>

```python
import gym

env = gym.make('CartPole-v0')
print(env.action_space)
print(env.observation_space)
```

```python
print(env.observation_space.high)
print(env.observation_space.low)
```

```python
#sample from a space or check something belongs to it.
from gym import spaces
space = spaces.Discrete(8) # Set with 8 elements {0, 1, 2, ..., 7}
x = space.sample()
assert space.contains(x)
assert space.n == 8
```

## Videos in Gym

```python
import gym
from gym import wrappers
path = 'vid'
env = gym.make('CartPole-v0')
env = wrappers.Monitor(env, path)

for i_episode in range(2):
        observation = env.reset()
        t = 0
        while True:
                env.render()
                print(observation)
                action = env.action_space.sample()
                observation, reward, done, info = env.step(action)
                t += 1
                if done:
                       print("Done after {} steps".format(t+1))
                       break
env.close()
```
